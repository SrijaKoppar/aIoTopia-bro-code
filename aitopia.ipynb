```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
#using fc for fully connected nodes of the layers 1 and 2 may add more layers if necessary and may change the number of nodes if necessary
#feeding 3 inputs to the neural network ie temperature humdidity and windspeed 

class Model(nn.Module):
    
    def _init_(self,in_features=3,h1=8,h2=8,outfeatures=3):
        super()._innit_() #calling the super from the parent class which is inbuilt
        self.fc1 = nn.linear(in_features,h1)
        self.fc2 = nn.linear(h1,h2)
        self.out = nn.linear(h2,out_features)

def forward(self,x):
    x = F.relu(self.fc1(x))
    x = F.relu(self.fc2(x))
    x = self.out(x)

    return x
```

```python
#creating a manual seed 
torch.manual_seed(45)
#creating object
model = Model()
```
